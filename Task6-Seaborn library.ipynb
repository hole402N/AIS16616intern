# Seaborn library

# 1. Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Total Bill vs. Size of the Group')
plt.xlabel('Size of the Group')
plt.ylabel('Total Bill')
plt.show()
```

# 2. Box Plot

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

# 3. Scatter Plot

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

# 4.Regression Plot

```python
# Regression plot
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs. Tip with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

# 5.Pair Plot (Grid Plot)

```python
# Pair plot
sns.pairplot(tips)
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```

# 6.Histogram

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=30, kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

# 7.Residual Plot

```python
# Custom Residual Plot
plt.figure(figsize=(10, 6))
sns.residplot(x='total_bill', y='tip', data=tips, color='purple', lowess=True)
plt.title('Customized Residual Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```

# 8.Count Plot with Multiple Variables

```python
# Count Plot with multiple variables
plt.figure(figsize=(10, 6))
sns.countplot(x='day', data=tips, hue='sex')
plt.title('Count Plot of Days with Gender')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
```
